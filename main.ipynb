#### Loop over the files and extract features

```python
import re

def count_text_length(text):
    # 加载 Stanza 的中文处理器
    #stanza.download('zh')
    #nlp = stanza.Pipeline('zh', processors='tokenize')
    
    # 读取文件内容
    #with open(file_path, 'r', encoding='utf-8') as f:
    #    text = f.read()
    
    # 运行 Stanza 分析
    doc = nlp(text)
    
    # 统计句子数和词数
    num_sents = len(doc.sentences)
    num_words = sum(len(sentence.words) for sentence in doc.sentences)
    
    # 统计汉字数（不包括标点符号）
    num_chars = len(re.findall(r'[一-鿿]', text))
    
    # 输出结果
    print(f"Number of sentences: {num_sents}")
    print(f"Number of words: {num_words}")
    print(f"Number of Chinese characters: {num_chars}")
    
    return num_sents, num_words, num_chars
```

```python
features = []
subjects = []
lengths = []

for file in os.listdir(folder):
    print("Processing file: ", file)
    file_path = os.path.join(folder, file)
    text = open(file=file_path, encoding='utf-8').read()
    doc = nlp(text)

    num_sents, num_words, num_chars = count_text_length(text)
    
    #num_sentences, num_clauses, num_words, ratio_unique_words, ratio_repeats, mean_num_clauses_per_sentence, mean_num_t_units_per_sentence, mean_num_verb_phrases_per_t_unit, mean_num_clauses_per_t_unit, mean_length_sentences, mean_length_clauses, mean_length_t_units, prop_noun_phrase, prop_verb_phrase, prop_adj_phrase, prop_adv_phrase, prop_prep_phrase = get_counting_features(doc)
    #mean_depth, ratio_adv_mods, ratio_adj_mods, ratio_nom_mods, ratio_nom_subjs, ratio_cases, ratio_compounds, mean_dep_dis, mean_hier_dis, ratio_head_init_deps, ratio_head_final_deps = get_struct_features(doc)
    subjects.append(file)
    #features.append([num_sentences, num_clauses, num_words, ratio_unique_words, ratio_repeats, mean_num_clauses_per_sentence, mean_num_t_units_per_sentence, mean_num_verb_phrases_per_t_unit, mean_num_clauses_per_t_unit, mean_length_sentences, mean_length_clauses, mean_length_t_units, prop_noun_phrase, prop_verb_phrase, prop_adj_phrase, prop_adv_phrase, prop_prep_phrase, mean_depth, ratio_adv_mods, ratio_adj_mods, ratio_nom_mods, ratio_nom_subjs, ratio_cases, ratio_compounds, mean_dep_dis, mean_hier_dis, ratio_head_init_deps, ratio_head_final_deps])
    lengths.append([num_sents, num_words, num_chars])
```
